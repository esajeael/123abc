```python
year=GDP["Year"]
```

```python
def make_dashboard(x, gdp_change, unemployment, title, file_name):
    output_file(file_name)
    p = figure(title=title, x_axis_label='year', y_axis_label='%')
    p.line(x.squeeze(), gdp_change.squeeze(), color="firebrick", line_width=2, legend="% GDP change")
    p.line(x.squeeze(), unemployment.squeeze(), line_width=2, legend="% unemployed")
    show(p)
```

```python
make_dashboard(year,GDP['GDP'],EMP['Annual'],'GDP-Unemployement Comparison\nUSA\n1961-2018',"Result")
```
